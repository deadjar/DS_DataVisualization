```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = r'C:\\Users\\jaded\\Downloads'
os.chdir(path)
df = pd.read_csv('Large_NYC_Visualization_Dataset.csv',sep = ',')
df.head()
```

```python
df.columns
df.head(30)
```

# First Visualization

```python
neighborhood_average_prices = df.groupby('Neighborhood')['House_Prices_Median'].median().reset_index()
neighborhood_average_prices.rename(columns={'Neighborhood':'Borough'})
```

```python
sns.barplot(x='Neighborhood',y='House_Prices_Median',data=neighborhood_average_prices, width=0.5)
plt.title('Average Median House Prices In Each Borough')
plt.xlabel('Borough')
plt.ylabel('Average Median Prices')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
sns.set_palette("Set3")
plt.show()
```

# Second Visualization

```python
neighborhood_population = df.groupby('Neighborhood')['Population_Thousands'].sum().reset_index()
neighborhood_population.head()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(neighborhood_population['Population_Thousands'], labels=neighborhood_population['Neighborhood'],autopct='%1.1f%%')
plt.title('Population Distribution by Neighborhood')
plt.show()
```
